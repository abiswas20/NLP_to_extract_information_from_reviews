```
reviews_ratings.dropna(inplace=True)
```

```
X=reviews_ratings['reviewText']
y=reviews_ratings['overall']
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y)
```

```
y_train=y_train[:500000]  #y values for the first 500,000 rows from training set.
y_test=y_test[:50000]  #y values for the first 50,000 rows from the test set.
```

```
y_train.to_csv('/content/drive/My Drive/amazon_reviews_project/y_train',index=0)
y_test.to_csv('/content/drive/My Drive/amazon_reviews_project/y_test',index=0)
```

```
del X
gc.collect()
```

```
def custom_tokenize(text):
    if not text:
        print('The text to be tokenized is a None type. Defaulting to blank string.')
        text = ''
    return word_tokenize(text,preserve_line=True)
```

```
#tokenized first 50 words from each review
X_train_tokenized=[]
list_of_indices=list(X_train.index)[:500000]   #only the first 500,000 indices.
for i in list_of_indices:
    X_train_tokenized.append(custom_tokenize(X_train[i])[:50])
```

```
X_train_tokenized
```

```
del reviews_ratings
del X_train
gc.collect()
```

```
ps=PorterStemmer()
X_train_tokenized_stemmed=[]
for k in X_train_tokenized:
    X_train_tokenized_stemmed.append([ps.stem(l) for l in k])
```

```
del X_train_tokenized
gc.collect()
```

```
def dummy_func(doc):
    return doc
```

```
tfidf = TfidfVectorizer(analyzer='word',tokenizer=dummy_func,ngram_range=(1,3),preprocessor=dummy_func,stop_words=stop_words,max_features=1500,token_pattern=None)  
```

```
X_train_vectorized=tfidf.fit_transform(X_train_tokenized_stemmed)
```

```
df_train_vectorized=pd.DataFrame(data=X_train_vectorized.toarray(),columns=tfidf.get_feature_names())
df_train_vectorized.to_csv('/content/drive/My Drive/amazon_reviews_project/vectorized_train_data.csv',index=0)  #saving vectorized representation of training data
```

```
#triggering garbage collection
del X_train_tokenized_stemmed
del df_train_vectorized
del y_train
del y_test
gc.collect()
```

```
#Variable check to improve garbage collection
import sys
def sizeof_fmt(num, suffix='B'):
    ''' by Fred Cirera,  https://stackoverflow.com/a/1094933/1870254, modified'''
    for unit in ['','Ki','Mi','Gi','Ti','Pi','Ei','Zi']:
        if abs(num) < 1024.0:
            return "%3.1f %s%s" % (num, unit, suffix)
        num /= 1024.0
    return "%.1f %s%s" % (num, 'Yi', suffix)

for name, size in sorted(((name, sys.getsizeof(value)) for name, value in locals().items()),
                         key= lambda x: -x[1])[:10]:
    print("{:>30}: {:>8}".format(name, sizeof_fmt(size)))
```
